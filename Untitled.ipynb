```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Load iPhone 15 CSV file
df = pd.read_csv("iphone15_sentiment_final.csv")

# Check columns
print("Columns in dataset:")
print(df.columns)

# Convert rating into True Label
# 4-5 = Positive
# 3 = Neutral
# 1-2 = Negative

def rating_to_label(rating):
    if rating >= 4:
        return "Positive"
    elif rating == 3:
        return "Neutral"
    else:
        return "Negative"

# Create Ground Truth column
df["True_Label"] = df["rating"].apply(rating_to_label)

# IMPORTANT FIX:
# Convert BERT labels from UPPERCASE to Proper format

df["BERT_Label"] = df["BERT_Label"].str.capitalize()

print("\nSample Labels:")
print(df[["rating", "True_Label", "TextBlob_Label", "VADER_Label", "BERT_Label"]].head())

# Correct model columns
models = {
    "TextBlob": "TextBlob_Label",
    "VADER": "VADER_Label",
    "BERT": "BERT_Label"
}

results = []

# Calculate Accuracy + F1 Score
for model_name, column in models.items():

    print(f"\n========== {model_name} ==========")

    accuracy = accuracy_score(
        df["True_Label"],
        df[column]
    )

    f1 = f1_score(
        df["True_Label"],
        df[column],
        average="weighted"
    )

    results.append([
        model_name,
        accuracy,
        f1
    ])

    print("Accuracy Score:", round(accuracy, 4))
    print("F1 Score:", round(f1, 4))

    print("\nClassification Report:")
    print(classification_report(
        df["True_Label"],
        df[column]
    ))

# Final comparison table
result_df = pd.DataFrame(
    results,
    columns=["Model", "Accuracy", "F1 Score"]
)

print("\n========== Final Comparison Table ==========")
print(result_df)

# Save output
result_df.to_csv(
    "iphone15_model_performance_comparison.csv",
    index=False
)

print("\nFile saved successfully:")
print("iphone15_model_performance_comparison.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load file
df = pd.read_csv("iphone15_sentiment_final.csv")

# Fix BERT labels
df["BERT_Label"] = df["BERT_Label"].str.capitalize()

# Select Positive and Negative reviews
positive_reviews = df[df["BERT_Label"] == "Positive"]["review_text"].dropna()
negative_reviews = df[df["BERT_Label"] == "Negative"]["review_text"].dropna()

# Convert reviews into single text
positive_text = " ".join(positive_reviews.astype(str))
negative_text = " ".join(negative_reviews.astype(str))

# -----------------------------
# Positive Word Cloud + Save
# -----------------------------
positive_wc = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(positive_text)

plt.figure(figsize=(10,5))
plt.imshow(positive_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Positive Review Word Cloud")

# Save image file
plt.savefig("iphone15_positive_wordcloud.png")

plt.show()

# -----------------------------
# Negative Word Cloud + Save
# -----------------------------
negative_wc = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(negative_text)

plt.figure(figsize=(10,5))
plt.imshow(negative_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Negative Review Word Cloud")

# Save image file
plt.savefig("iphone15_negative_wordcloud.png")

plt.show()

print("Word clouds saved successfully!")
print("Files created:")
print("1. iphone15_positive_wordcloud.png")
print("2. iphone15_negative_wordcloud.png")
```

```python
# =========================================================
# Final Comparison of All 3 Mobiles + Actual Ratings
# iPhone 15 + iPhone 16 + iQOO Z10
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# STEP 1: Load Performance Comparison Files
# ---------------------------------------------------------

iphone15 = pd.read_csv("iphone15_model_performance_comparison.csv")
iphone16 = pd.read_csv("iphone16_model_performance_comparison.csv")
iqoo = pd.read_csv("iqoo_z10_model_performance_comparison.csv")

# Add Mobile Name column
iphone15["Mobile"] = "iPhone 15"
iphone16["Mobile"] = "iPhone 16"
iqoo["Mobile"] = "iQOO Z10"

# Combine all model performance tables
final_comparison = pd.concat(
    [iphone15, iphone16, iqoo],
    ignore_index=True
)

print("========== Combined Model Comparison ==========")
print(final_comparison)

# Save final combined comparison
final_comparison.to_csv(
    "all_mobile_model_comparison.csv",
    index=False
)

# ---------------------------------------------------------
# STEP 2: Load Original Files for Actual Rating Comparison
# ---------------------------------------------------------

df1 = pd.read_csv("iphone15_sentiment_final.csv")
df2 = pd.read_csv("iphone16_sentiment_final.csv")
df3 = pd.read_csv("iqoo_z10_sentiment_final.csv")

# Calculate Average Actual Rating
actual_rating_data = pd.DataFrame({
    "Mobile": [
        "iPhone 15",
        "iPhone 16",
        "iQOO Z10"
    ],
    "Average_Actual_Rating": [
        df1["rating"].mean(),
        df2["rating"].mean(),
        df3["rating"].mean()
    ]
})

print("\n========== Actual Rating Comparison ==========")
print(actual_rating_data)

# Save actual rating comparison
actual_rating_data.to_csv(
    "actual_rating_comparison.csv",
    index=False
)

# ---------------------------------------------------------
# STEP 3: Graph for Model Comparison (All Mobiles)
# ---------------------------------------------------------

plt.figure(figsize=(10,6))

for model in final_comparison["Model"].unique():
    temp = final_comparison[final_comparison["Model"] == model]
    plt.plot(
        temp["Mobile"],
        temp["Accuracy"],
        marker="o",
        label=model
    )

plt.title("Accuracy Comparison of Models Across Mobiles")
plt.xlabel("Mobile")
plt.ylabel("Accuracy Score")
plt.legend()

plt.savefig("all_mobile_model_accuracy_comparison.png")
plt.show()

# ---------------------------------------------------------
# STEP 4: Graph for Actual Rating Comparison
# ---------------------------------------------------------

plt.figure(figsize=(8,5))

plt.bar(
    actual_rating_data["Mobile"],
    actual_rating_data["Average_Actual_Rating"]
)

plt.title("Average Actual Rating Comparison")
plt.xlabel("Mobile")
plt.ylabel("Average Rating")

plt.savefig("actual_rating_comparison_graph.png")
plt.show()

# ---------------------------------------------------------
# FINAL OUTPUT
# ---------------------------------------------------------

print("\nFiles saved successfully:")
print("1. all_mobile_model_comparison.csv")
print("2. actual_rating_comparison.csv")
print("3. all_mobile_model_accuracy_comparison.png")
print("4. actual_rating_comparison_graph.png")
```

```python
import pandas as pd

# Load all three sentiment files
iphone15 = pd.read_csv("iphone15_sentiment_final.csv")
iphone16 = pd.read_csv("iphone16_sentiment_final.csv")
iqoo = pd.read_csv("iqoo_z10_sentiment_final.csv")

datasets = {
    "iPhone 15": iphone15,
    "iPhone 16": iphone16,
    "iQOO Z10": iqoo
}

# Convert sentiment labels to numeric scores
def label_to_score(label):
    label = str(label).strip().upper()

    if "POS" in label:
        return 5
    elif "NEG" in label:
        return 1
    else:
        return 3

# Build result table
results = []

for phone, df in datasets.items():
    actual = df["rating"].mean()

    df["tb_score"] = df["TextBlob_Label"].apply(label_to_score)
    df["vader_score"] = df["VADER_Label"].apply(label_to_score)
    df["bert_score"] = df["BERT_Label"].apply(label_to_score)

    # Average of all 3 model scores
    cumulative = df[
        ["tb_score", "vader_score", "bert_score"]
    ].mean(axis=1).mean()

    diff = cumulative - actual

    results.append({
        "Mobile": phone,
        "Actual Rating": round(actual, 4),
        "Cumulative Rating": round(cumulative, 4),
        "Difference": round(diff, 4)
    })

# Final table
result_df = pd.DataFrame(results)

print(result_df.to_string(index=False))

# =====================================
# Save into CSV file
# =====================================

result_df.to_csv(
    "final_mobile_rating_comparison.csv",
    index=False
)

print("\nFile saved successfully!")
print("Saved as: final_mobile_rating_comparison.csv")
```
